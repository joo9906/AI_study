```python
import json
from pathlib import Path

path = Path("mclm_k2b_32.jsonl")  # change to your file
records = []

with path.open("r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:      # skip blank lines
            continue
        records.append(json.loads(line))
```

```python
prompt = """You will be given a question and 4 responses. Return the index of the best response.
No explanation, return the index only in [[N]] format.
DO NOT SOLVE THE QUESTION RETURN THE INDEX OF THE BEST ANSWER.

Answer are provided in 
----
[idx] {{content}}
----

format.

Question : {question}

Responses : """


for row in records:
    query = prompt.format(question=row['question'])
    query += '\n'
    query +='\n\n'.join(['-'*20 +'\n'+ f"[{idx}] {resp}" + '\n'+'-'*20  for idx,resp in enumerate(row['responses'][:4])])

    response = batch_completion_models(
        models=[
        "openrouter/z-ai/glm-4.5-air:free",
        'openrouter/meituan/longcat-flash-chat:free',
        'openrouter/z-ai/glm-4.5-air',
        'openrouter/moonshotai/kimi-k2:free',
        'openrouter/mistralai/mistral-small-3.2-24b-instruct:free',
        'openrouter/meta-llama/llama-4-scout:free',
        'openrouter/meta-llama/llama-4-maverick:free',
        'openrouter/deepseek/deepseek-chat-v3-0324:free'
        ], 
        messages=[{"role": "user", "content": query}]
    )

    break
```
