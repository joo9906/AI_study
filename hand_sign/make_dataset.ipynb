```python
import os

os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
# 필요한 라이브러리 로드

import re
import zipfile

from extract_coordinate import process_with_view
```

```python
# video_zip 파일 경로

video_zip_path = "/home/j-j13a602/raw_video_zip"
```

```python
# 정면 mp4 파일 받을 리스트
front_files = []
```

```python
# zip 파일 개수 찾기

folder_list = [
    os.path.join(video_zip_path, f)
    for f in os.listdir(video_zip_path)
    if re.search(r"real", f, re.IGNORECASE)
]
folder_list
```

```python
# mp4 파일 찾기

front_files += [
    name for name in os.listdir(folder_list[0])
    if re.search(f"_F.mp4$", name, re.IGNORECASE)
]
front_files[:3]
```

```python
# 좌표 추출한 csv 저장할 폴더 설정

out_root = "/home/j-j13a602/make_dataset/annotation"

os.makedirs(out_root, exist_ok=True)
```

```python
# 좌표 추출하기 - REAL

for i, video_file in enumerate(front_files):
    video_zip_path = folder_list[0]
    video_path = os.path.join(video_zip_path, video_file)
    main_name = video_file.strip(".mp4")
    word_id = int(video_file.split("WORD")[-1][:4])
    person_id = int(video_file.split("REAL")[-1][:2])
    out_coords = os.path.join(out_root, f"{main_name}.csv")  # CSV 또는 .json

    result = process_with_view(
        video_path=video_path,
        out_coords_path=out_coords,
        out_format="csv",          # "json"로 바꾸면 줄단위 JSON
        every_n=1,                 # 프레임 샘플링(2~3으로 올리면 속도↑)
        resize_factor=1.0,         # 손이 작게 보이면 1.5~2.0
        fps_override=30,
        word_id=word_id,
        person_id=person_id
    )
```
